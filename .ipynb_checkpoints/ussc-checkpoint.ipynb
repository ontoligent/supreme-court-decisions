```python
import pandas as pd
```

# Import CSV

```python
csv = pd.read_csv('ussc_data.csv')
```

# Format columns

```python
csv.columns = csv.columns.str.lower().str.replace(' ', '_')
```

```python
csv.columns
```

# Define ID

```python
csv['case_id'] = csv.filename.apply(lambda x: x.split('.')[0])
```

# Get OPINION columns

## Get concurring opinion cols

```python
concur_cols = [x for x in csv.columns[csv.columns.str.match('^concur_\d$')]]
```

```python
concur_auth_cols = [x for x in csv.columns[csv.columns.str.match('^concur_\d_.+$')]]
```

```python
# concur_cols
```

```python
# concur_auth_cols
```

## Get dissenting opinion cols

```python
dissent_cols = [x for x in csv.columns[csv.columns.str.match('^dissent_\d$')]]
```

```python
# dissent_cols
```

```python
dissent_auth_cols = [x for x in csv.columns[csv.columns.str.match('^dissent_\d_.+$')]]
```

```python
# dissent_auth_cols
```

# Create CASE table

```python
case_cols = set(csv.columns) - set(concur_cols) - set(dissent_cols) - set(concur_auth_cols) - set(dissent_auth_cols) - {'opinion'}
```

```python
case_cols
```

```python
CASE = csv[case_cols].set_index('case_id')
```

```python
CASE.head()
```

```python
CASE = CASE.drop(['court','index','opinion_author','filename'], 1)
```

```python
CASE
```

# Create CORPUS table

```python
dfs = []

dfs.append(csv[['year', 'case_id', 'opinion_author', 'opinion']].copy())
dfs[0]['position'] = 'opinion'
dfs[0] = dfs[0].rename(columns={'opinion_author':'author_id'})

for x, y in zip(concur_auth_cols, concur_cols):
    df = csv[['year','case_id', x, y]].copy()
    df = df.rename(columns={x:'author_id', y:'opinion'})
    df['position'] = 'concur'
    df = df.dropna()
    dfs.append(df)

for x, y in zip(dissent_auth_cols, dissent_cols):
    df = csv[['year','case_id', x, y]].copy()
    df = df.rename(columns={x:'author_id', y:'opinion'})
    df['position'] = 'dissent'
    df = df.dropna()
    dfs.append(df)
    
CORPUS = pd.concat(dfs).sort_index()
```

```python
corpus = csv[corpus_cols].set_index(['year','case_id']).stack().to_frame('doc_content').sort_index().reset_index()
corpus['position'] = corpus.level_2.apply(lambda x: x.split('_')[0])
corpus['position_ord'] = corpus[corpus.position.str.match(r'^(con|dis)')].level_2.apply(lambda x: x.split('_')[-1])
corpus.loc[corpus.position_ord.isna(), 'position_ord'] = 1
corpus = corpus.set_index(['year','case_id','position','position_ord'])[['doc_content']]
```

```python
corpus
```

# Consolidate by position

```python
M = corpus.groupby(['year','case_id','position']).doc_content.apply(lambda x: ' '.join(x)).to_frame().unstack()
M.columns = M.columns.droplevel(0)
```

# Keep only those with all three

```python
M2 = M.dropna().stack().to_frame('doc_content')
```

```python
# M2 = M.dropna(subset=['dissent','concur']).stack().to_frame('doc_content')
```

```python
M2['doc_len'] = M2.doc_content.str.len()
```
