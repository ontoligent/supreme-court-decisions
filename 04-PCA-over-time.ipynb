# PCA over Time

# Set Up

```python
import pandas as pd
import numpy as np
from lib.helper import Db, lmplot
```

```python
import seaborn as sns
sns.set()
sns.set_theme('poster')
```

# Import Tables

```python
base_path = './'
db_file = f'{base_path}db/ussc.db'
db = Db(db_file)
OHCO = ['vol_num','case_num','position']
```

```python
db.import_table('CASE', OHCO[:2])
db.import_table('YEAR', 'year')
db.import_table('PCA_THETA', OHCO)
db.import_table('PCA_TOPICS', 'topic_id')
```

# Inspect Components

```python
db.PCA_THETA.var().sort_values().plot.barh(figsize=(2.5,5));
```

```python
_ = db.PCA_TOPICS.apply(lambda x: print(x.name, '\n + ', x.topwords.replace('|', '\n - ')), 1)
```

# Add Year to THETA

```python
X1 = db.PCA_THETA.join(db.CASE.year, on=['vol_num','case_num']).reset_index().set_index(OHCO + ['year'])
```

```python
X1.head()
```

```python
X2 = X1.stack().to_frame('topic_weight')
X2.index.names = OHCO + ['year', 'topic_id']
```

```python
# X2.reset_index().head()
```

# Mean Weights by Year

```python
X3 = X2.groupby(['year','topic_id']).topic_weight.sum().unstack()
```

```python
X3
```

```python
X4 = X3 #[X3.index >= 1900]
```

# Plot Components over Time

```python
for i in range(10):
    col = f'PC{i}'
    topwords = db.PCA_TOPICS.loc[col].topwords.replace('|', '\n - ')
    sns.relplot(data=X4, x='year', y=col, 
                hue=db.YEAR.opinion_count,
                size=db.YEAR.opinion_count,
                sizes=(0, 500),
                height=7.5,
                aspect=2.5).set(title=f'+ {topwords}')
```

# Explore Words

```python
db.import_table('PCA_PHI', 'term_str')
```

```python
for i in range(10):
    pc = f'PC{i}'
    neg = db.PCA_PHI[pc].sort_values().head(10)
    pos = db.PCA_PHI[pc].sort_values().tail(10)
    combo = pd.concat([neg, pos]).to_frame().reset_index()
    combo.plot.barh('term_str', pc, figsize=(5,10), title=pc, legend=False)
```

```python
pc_idx = (db.PCA_PHI[pc] <= db.PCA_PHI[pc].quantile(.01)) | (db.PCA_PHI[pc] > db.PCA_PHI[pc].quantile(.99))
```

```python
db.PCA_PHI.loc[pc_idx, [pc]].sort_values(pc).reset_index().plot.barh('term_str', pc, figsize=(10,30), legend=False)
```
